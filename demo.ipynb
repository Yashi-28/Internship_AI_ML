```python
import cv2
face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)

img = cv2.imread("face.jpg")   

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Detect faces
faces = face_cascade.detectMultiScale(
    gray,
    scaleFactor=1.1,
    minNeighbors=5,
    minSize=(30, 30)
)

# Count faces
print("Number of faces detected:", len(faces))

# Draw rectangles around faces
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)

# Display count on image
cv2.putText(img,
            f"Faces: {len(faces)}",
            (10, 30),
            cv2.FONT_HERSHEY_SIMPLEX,
            1,
            (0, 0, 255),
            2)

# Show image
cv2.imshow("Face Detection", img)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
import cv2

# Load Haar Cascade
face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)

# Read image
img = cv2.imread("face2.jpg")   # Replace with your image name

# Check if image is loaded
if img is None:
    print("Error: Image not found!")
    exit()

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Detect faces
faces = face_cascade.detectMultiScale(
    gray,
    scaleFactor=1.1,
    minNeighbors=4,
    minSize=(30, 30)
)

# Print number of faces
print("Number of faces detected:", len(faces))

# Draw rectangles
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)

# Display count on image
cv2.putText(
    img,
    f"Faces: {len(faces)}",
    (10, 30),
    cv2.FONT_HERSHEY_SIMPLEX,
    1,
    (0, 0, 255),
    2
)

# Show image
cv2.imshow("Face Detection", img)

cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
import cv2

model = YOLO("yolov8n.pt")

results = model("car.jpg", conf=0.25)

annotated_img = results[0].plot()

car_count = 0

for box in results[0].boxes:
    cls = int(box.cls[0])

    if cls == 2:      # Car class
        car_count += 1

print("Total Cars:", car_count)

cv2.imshow("Car Detection", annotated_img)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
import cv2

# Load YOLO model
model = YOLO("yolov8n.pt")

# Image path
img_path = "dogs.jpg"   

# Detect only dogs (class ID = 16)
results = model(img_path, classes=[16], conf=0.25)

# Get annotated image
annotated_img = results[0].plot()

# Count dogs
dog_count = len(results[0].boxes)

print("Total Dogs Detected:", dog_count)

# Save output image
cv2.imwrite("detected_dogs.jpg", annotated_img)

# Show output
cv2.imshow("Dog Detection", annotated_img)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```
